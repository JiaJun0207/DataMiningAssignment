Q1: Do individuals with high marital satisfaction tend to report both frequent communication and mutual decision-making?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations
from collections import defaultdict
```

Load dataset

```python
df = pd.read_csv("Marital-satisfaction_Data.csv")
```

```python
# --- 1. Data Exploration (statistics and visualization) ---
q1_cols = ['KMSS1', 'KMSS2', 'KMSS3', 'Scale1', 'Scale2']
df_q1 = df[q1_cols].dropna()

# Create categoricals for plotting
df_q1['Satisfaction_High'] = df_q1['KMSS1'] >= 4

df_q1['Communication_Frequent'] = df_q1['Scale1'] >= 4
df_q1['Decision_Making_Shared'] = df_q1['Scale2'] >= 4

# Bar chart: count of Satisfaction High vs Communication/Decision Making
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='Satisfaction_High', hue='Communication_Frequent', data=df_q1)
plt.title("Communication vs High Satisfaction")

plt.subplot(1, 2, 2)
sns.countplot(x='Satisfaction_High', hue='Decision_Making_Shared', data=df_q1)
plt.title("Decision-Making vs High Satisfaction")

plt.tight_layout()
plt.show()
```
